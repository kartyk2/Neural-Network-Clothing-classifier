```python
# Import dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load training dataset

train_df= pd.read_csv('fashion-mnist_train.csv') 
```

```python
# map the labels
labels= {
    0: "T-shirt/top",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle boot"
}
```

```python
# see the fashion item in the dataset by index
def get_image(image_index: int):

    image= train_df[image_index: image_index+1].to_numpy()
    
    label= image[0][0]
    image= image[0][1:]

    reshaped_image= image.reshape(28, 28)

    print(labels[label])
    plt.imshow(reshaped_image, cmap='gray_r')
    plt.show()


for index in range(5):
    get_image(index)
```
